```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib notebook
```

Define the dy/df function

```python
def f(y, t, params):
    theta, omega = y      # unpack current values of y
    g = params[0]
    k = params[1] # unpack parameters
    derivs = [omega, g*np.sin(theta)+k*np.random.rand()]
    return derivs
# Just trying out this force
# f([np.pi/2,0], 0, 1)
```

```python
# Parameters

g = 1.0
kp = .0

# Initial values
theta0 = np.pi/2.     # initial angular displacement
omega0 = 0.0     # initial angular velocity

# Bundle parameters for ODE solver
params = [g,kp]

# Bundle initial conditions for ODE solver
y0 = [theta0, omega0]

# Make time array for solution
tStop = 20.
tInc = .001
k = kp/tInc
t = np.arange(0., tStop, tInc)
```

```python
# Call the Eulers method solver
trange = np.arange(0,tStop,tInc) 

thetan = theta0
omegan = omega0
thetasoln = np.zeros(len(trange))
omegasoln = np.zeros(len(trange))
for i in range(len(trange)): 
    thetanext = thetan + tInc*omegan
    #omeganext = thetan + tInc*(g*np.sin(thetan)+k*(np.random.rand()-.5))
    omeganext = omegan + tInc*g*np.sin(thetan)
    thetasoln[i]=thetanext
    omegasoln[i]=omeganext
    thetan = thetanext
    omegan = omeganext
    #print (thetan, omegan)
```

```python
# Plot results


# Plot theta as a function of time\
plt.figure()
plt.plot(trange, thetasoln)

# Plot omega as a function of 
plt.figure()
plt.plot(trange, omegasoln)

```
